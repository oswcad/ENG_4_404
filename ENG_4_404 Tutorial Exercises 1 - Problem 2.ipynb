## About the voltage $v_o$ 

- Notice the voltage $v_o$ is the same voltage across the resistor with value $80 \Omega$
- Notice that the resistors with values $30 \Omega$ and $90 \Omega$ are in series, so they can be replaced by a single resitor of $120 \Omega$
- Notice that the resistor of $80 \Omega$ is in parallel with the new found resistor of $120 \Omega$. 
- Notice that all the three resistors can be replaced by a single resistor. say $R_{EQ}$ like:

$$ R_{EQ} = 80 \vert \vert (30 + 90) = 80 \vert \vert 90 $$

**Note:** $\vert \vert $ is a way to indicate in calculation that left (of souble vertical bars) is in parallel with right. 

With the shorcut we know about two parallel resitors: $R_{EQ} = \frac{80(120)}{80 + 120} = 48 \Omega$

Then, since across this global equivalent resistor the current of 4A will circulate, then $v_o = 4R_{EQ} = 4(48) = 192 V$. 

## About the current $i_1$ 

- As we now know the voltage $v_o$ and that voltage is across resistor with value $80 \Omega$ then we also know $192 = 80i_1$, thus $i_1 = 192/80 = 2.4 A$. 

### Another way

- We can apply the current divider method directly, this way:

$$i_1 = \frac{120}{120 + 80}4 = \frac{120}{200}4 = 0.6(4) = 2.4 A$$

## About the current $i_2$ 

- As the voltage $v_o$ is also across the equivalent resistor with value $30 + 90 = 120 \Omega$ then we also know $192 = 120i_2$, thus $i_2 = 192/120 = 1.6 A$. 

### Another way

- We can apply the current divider method directly, this way:

$$i_2 = \frac{80}{120 + 80}4 = \frac{80}{200}4 = 0.4(4) = 1.6 A$$

$$P_{4A} = 192(4) = 768 W $$
