## About the equivalent resistance

- Notice the $6 \Omega$ resistor is in parallel with the resistor of value $4 \Omega$, thus all they can be replaced by $\frac{4(6)}{4 + 6} = \frac{768}{64} = 2.4 \Omega$
- Notice the $2.4 \Omega$ is in series with the resistor with value $1.6 \Omega$ thus all so far can be replaced by a single resistor of value $2.4 + 1.6 = 4 \Omega$
- Notice the $4 \Omega$ resistor is in parallel with the resistor of value $16 \Omega$, thus all they can be replaced by $\frac{4(16)}{4 + 16} = \frac{64}{20} = 3.2 \Omega$

*Answer*: The equivalent total resistance for the circuit is of $3.2 \Omega$

# About the voltage across the current source of 10A

The voltage is easy since $V_{10A} = 3.2(10) = 32 V$. 

# About the power of the current source

Notice the source is generating power and so the power is $P_{10A} = 32 V (10A) = 320 W$
