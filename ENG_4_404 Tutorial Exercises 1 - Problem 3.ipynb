3) Find the power developed by the 50V source

As power is voltage times current, we need to find out the current going across the voltage source. For this we can pick any point, either the point at the positive terminal of the voltage source (down in the figure) or at the negative terminal (up in the figure). We know that the sum of currents on any one of those points has to be zero. 

- We notice that there is a current circulating across the three resistors on the left of the voltage source
- We notice that there is a current circulating across the two resistors on the right of the voltage source
- There is a current circulating across the voltage source itself. 

*Equivalent resistance of the three resitors on the left*

These three resitors are in series and so the equivalenet resistance has the value $60 + 50 + 40 = 150 \Omega$

*Equivalent resistance of the two resitors on the right*
These two resitors are in series and so the equivalenet resistance has the value $30 + 45 = 75 \Omega$

## Sum of currents on the negative terminal of the voltage source

Let's call the current across the resistors on the left $I_1$. Let's call the current across the resistors on the right $I_2$. Let's call the current across the voltage source $I_T$. 

Let's put our attention on the point at the negative terminal of the voltage source. Like this

<img src="E13b.PNG"> 



Thus the sum of currents at the bottom point (more appropriate to call it *node*)

$I_1 + I_2 = I_T$; that is the sum of *leaving* currents from node equal to the sum of *incoming* currents to node. 

That was easy. Now we need to calacuate the values of currents $I_1$ and $I_2$.

*Calculating current $I_1$*: <br>

$I_1 = \frac{50}{150} = \frac{1}{3}$ A. It is ok to say $I_1 = 0.34$ A.

*Calculating current $I_2$*: <br>

$I_2 = \frac{50}{75} = \frac{2(25)}{3(25)} = \frac{2}{3}$ A. It is ok to say $I_2 = 0.66$ A.

*Calculating current $I_T$*: <br>

From above $I_T = I_1 + I_2 = \frac{1}{3} + \frac{2}{3} = \frac{3}{3} = 1$ A. 

Note this is the same as: $I_T = I_1 + I_2 = 0.34 + 0.66 = 1$ A. 

*Another way*: We could have calculated the parallel of resistors with values $150 \Omega$ and $75 \Omega$ and then from this the current $\frac{50}{150 \vert \vert 75}$ would be the same as $I_T$. Like so;

$R_T = \frac{75(150)}{75 + 150} = \frac{11250}{225} = 50 \Omega$. Thus $I_T = \frac{50}{50} = 1 A$. 


## Power of the voltage source 

Note that the current $I_T$ emerges from the positive terminal of voltage source and so the power is $p = vi$ or generating power.

$$P_{50V} = 50 V (1 A) = 50 W$$
