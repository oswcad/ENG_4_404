## About the resistance across terminals *a* and *b* 

- Notice that resistors with values $6 \Omega$ and $10 \Omega$ are in series, thus they can be replaced by a resistor of $16 \Omega$
- Notice the $16 \Omega$ resistor is in parallel with the resistor of value $48 \Omega$, thus all (so far) can be replaced by $\frac{16(48)}{16 + 48} = \frac{768}{64} = 12 \Omega$
- Notice the $12 \Omega$ is in series with the resistor with value $18 \Omega$ thus all so far can be replaced by a single resistor of value $12 + 18 = 30 \Omega$
- Notice the $30 \Omega$ resistor is in parallel with the resistor of value $15 \Omega$, thus all can be replaced by $\frac{30(15)}{30 + 15} = \frac{450}{45} = 10 \Omega$

*Answer*: The equivalent resistance across terminals *a* and *b* is of $10 \Omega$

### Voltage across the resisto of $48 \Omega$ 

From the diagram we know that $2A = I_1 + I_2$. From the equivalent resistance we calculated earlier we know that the resistor of $15 \Omega$ is in parallel with a resitance equivalent of $30 \Omega$. So we can apply the current divider principle to say <br>

$$I_1 = \frac{30}{30 + 15} 2 A = \frac{2(15)}{3(15)}2 A = \frac{2}{3}2 A = \frac{4}{3} A$$ 

Then the voltage across $15 \Omega$ resistor is of $\frac{4}{3}15 = 4(5) = 20 V$. 

for current $I_2$ things are simpler since $I_2 = 2 - \frac{4}{3} = \frac{6}{3} - \frac{4}{3} = \frac{2}{3} A$

The voltage across resistor with value $18 \Omega $ is of $\frac{2}{3}18 = 2(6) = 12 V$.  

Finally the voltage at the top of resistor with value $48 \Omega$ is of 12 V *less than* the voltage at the top of the resistor with value of $15 \Omega$ and of 20 - 12 = 8V. 

This is what we needed to show.
